# Climate Change and Food Scarcity in Developing Regions Like India and China

## Importing the Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## The first data set we look at is Annual Temperature Dataset


```python
global_temp = pd.read_csv("/Users/chinmaysalgaonkar/Desktop/temp_dataset.csv")
print(global_temp.head(10))
```

## We filter out data to get only recrods of NASA GISTEMP Source

```python
global_temp = global_temp[global_temp['Source'] == 'GISTEMP'].reset_index()[["Source", "Year", "Mean"]]
print(global_temp.head(10))
```

## We Plot Mean annual temperature vs Time (Years) 

```python

sns.set()
sns.lineplot(global_temp['Year'],global_temp['Mean'])
plt.ylabel("Mean")
plt.xlabel("Year")
plt.title("Average global mean temperature")
```

## Next We look at the Global Anuual Production of Wheat and Rice Yeild

### Starting by Reading the Dataset for Production of Crops Worldwide 

### 1) Wheat ( Yield in India and China )

```python
df_wheat = pd.read_csv("/Users/chinmaysalgaonkar/Desktop/Dataset1.csv")
print(df_wheat.head(5))
```
